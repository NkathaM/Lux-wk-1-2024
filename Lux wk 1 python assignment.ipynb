```python
# 1. Find all records where the weather was exactly clear.
clear_weather = df[df['Weather'] == 'Clear']
print("Records where the weather was exactly clear:")
print(clear_weather)
```

```python
# 2. Find the number of times the wind speed was exactly 4 km/hr.
wind_speed_4 = df[df['Wind Speed_km/h'] == 4].shape[0]
print("Number of times the wind speed was exactly 4 km/hr:", wind_speed_4)
```

```python
# 3. Check if there are any NULL values present in the dataset.
null_values = df.isnull().sum().sum()
print("Number of NULL values in the dataset:", null_values)
```

```python
# 4. Rename the column "Weather" to "Weather_Condition."
df.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)
print("Renamed column 'Weather' to 'Weather_Condition'.")
```

```python
# 5. Calculate the mean visibility of the dataset.
mean_visibility = df['Visibility_km'].mean()
print("Mean visibility of the dataset:", mean_visibility)
```

```python
# 6. Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.
wind_speed_greater_24_visibility_25 = df[(df['Wind Speed_km/h'] > 24) & (df['Visibility_km'] == 25)].shape[0]
print("Number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km:", wind_speed_greater_24_visibility_25)
```

```python
# 7. Calculate the mean value of each column for each weather condition.
mean_values_per_weather = df.groupby('Weather_Condition').mean()
print("Mean value of each column for each weather condition:")
print(mean_values_per_weather)
```

```python
# 8. Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.
specific_conditions = df[(df['Weather_Condition'] == 'Clear') & (df['Rel Hum_%'] > 50) | (df['Visibility_km'] > 40)]
print("Instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40:")
print(specific_conditions)
```

```python
# 9. Find the number of weather conditions that include snow.
snow_conditions = df[df['Weather_Condition'].str.contains('Snow', case=False)].shape[0]
print("Number of weather conditions that include snow:", snow_conditions)
```
